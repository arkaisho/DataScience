```python
import re
import pandas as pd
import numpy as np
import random
```

```python
#METHODS
def toUniqueList(serie):
    uniqueList=[]
    for row in serie:
        uniqueList+=row
    return pd.DataFrame(uniqueList)
def countDataframe(serie):
    return pd.DataFrame(toUniqueList(serie).value_counts(),columns=["count"]).reset_index().rename(columns={0:"item"})
def random_date(start, end, n=1):
    start_u = start.value//10**9
    end_u = end.value//10**9
    return pd.to_datetime(np.random.randint(start_u, end_u, n), unit='s')
```

```python
#READ LOREM IPSUM LINES
lorem_ipsum_source = "loremIpsum.txt"
lipsum_lines = []
with open(lorem_ipsum_source,encoding='utf-8') as file:
    line = file.readline() #read fisrt line
    while line:
        lipsum_lines.append(line.strip())
        line = file.readline()
#CREATE DATAFRAME WITH LINES
fake_data = pd.DataFrame(lipsum_lines)
fake_data = fake_data.loc[fake_data[0]!=''].reset_index(drop=True).rename(columns={0:"line"})
#EXTRACT WORDS FROM LINES
word_pattern = r"[\w.,]+"
fake_data['word']=fake_data.line.str.findall(word_pattern)
fake_words = toUniqueList(fake_data.word).copy().rename(columns={0:"word"})
#USE FAKE WORDS TO CREATE A DATASET OF FAKE MESSAGES
i=0
end=len(fake_words)
fake_phrases=[]
while i<end:
    r = random.randint(1,10)
    if i+r<end:
        fake_phrases.append(' '.join(fake_words.word[i:i+r]))
    i+= r
fake_data = pd.DataFrame(fake_phrases,columns=['message'])
fake_data
```

```python
#CREATE A LIST OF FAKE PERSONS
fake_persons= ['person'+str(i) for i in range(10)]
fake_data['author'] = np.random.choice(fake_persons,fake_data.shape[0])
fake_data
```

```python
#RANDOMLY APPEND EMOJIS TO MESSAGES
fake_emojis=str('😀 😃 😄 😁 😆 😅 😂 🤣 ☺️ 😊 😇 🙂 🙃 😉 😌 😍 🥰 😘 😗 😙 😚 😋 😛 😝 😜 🤪 🤨 🧐 🤓 😎 🤩 🥳 😏 😒 😞 😔 😟 😕 🙁 ☹️ 😣 😖 😫 😩 🥺 😢 😭 😤 😠 😡 🤬 🤯 😳 🥵 🥶 😱 😨 😰 😥 😓 🤗 🤔 🤭 🤫 🤥 😶 😐 😑 😬 🙄 😯 😦 😧 😮 😲 🥱 😴 🤤 😪 😵 🤐 🥴 🤢 🤮 🤧 😷 🤒 🤕 🤑 🤠 😈 👿 👹 👺 🤡 💩 👻 💀 ☠️ 👽 👾 🤖 🎃 😺 😸 😹 😻 😼 😽 🙀 😿 😾 ').split(' ')
def addEmoji(message):
    if random.randint(0,10)>7:
        return message+np.random.choice(fake_emojis)
    else:
        return message
fake_data.message = fake_data.message.apply(addEmoji)
```

```python
#RANDOM GIVE MESSAGES DATES
start = pd.to_datetime('2018-01-01')
end = pd.to_datetime('2020-01-01')
fake_data['date'] = random_date(start,end,n=len(fake_data))
fake_data.set_index('date').sort_index()
```

```python
#WRITE THE FAKE DATASET AS A WPP.TXT FILE
f = open("conversas.txt", "w")
for i,row in fake_data.iterrows():
    line = str(row.date).replace('-','/')[:-3]+' - '+row.author+': '+row.message+'\n'
    f.write(line)
f.close()
```
